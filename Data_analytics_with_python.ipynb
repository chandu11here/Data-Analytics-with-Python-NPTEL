Defining a Function for calculating alpha value

```python
def z_value(x,mu,SE):
  z=(x-mu)/SE;
  if z<0:
    alpha=stats.norm.cdf(z)
  else:
    alpha=1-stats.norm.cdf(z)
  print(alpha)
```

Calculating alpha for different values of x,mu and SE

```python
x=48.5
mu=50
SE=0.79
z_value(x,mu,SE)
```

Defining a Function for calculating beta value

```python
def type_2(mu1,mu2,sigma,n,alpha):
  z_tab=stats.norm.ppf(alpha)
  x_bar=mu1+(z_tab*(sigma/np.sqrt(n)))
  z_cal=(x_bar-mu2)/(sigma/np.sqrt(n))
  if(mu1<mu2):
    beta=1-stats.norm.cdf(z_cal)
  else:
    beta=stats.norm.cdf(z_cal)
  print(beta)
```

```python
type_2(8.3,7.4,3.1,60,0.05)
```

Hypothesis test about difference between two Population Proportion

When P1=P2=P

```python
def two_sample_proportion(P1,P2,n1,n2):
  p_pool=(P1*n1+P2*n2)/(n1+n2)
  se_pool=math.sqrt(p_pool*(1-p_pool)*(1/n1+1/n2))
  z=(P1-P2)/se_pool
  if (z<0) :
    p_value=stats.norm.cdf(z)
  else:
    p_value=1-stats.norm.cdf(z)
  return z, p_value*2

```

```python
print(two_sample_proportion(0.27,0.19,147,88))
```

Estimate b0,b1,standard Deviation of error using MLE

```python
from scipy.optimize import minimize
def lik(parameters):
  b1=parameters[0] #slope
  b0=parameters[1] #y-intercept
  sigma=parameters[2] #S.D. of error
  for i in np.arange(0,len(x)):
    y_exp=b0+b1*x
  L=len(x)/2*np.log(2*np.pi)+len(x)/2*np.log(sigma**2)+1/(2*sigma**2)*sum((y-y_exp)**2)
  return(L)

x=np.array([1,4,5,6,9])
y=np.array([2,6,7,9,15])
lik_model=minimize(lik,np.array([2,2,2]),method='Nelder-Mead')
print(lik_model.x)
```

```python
#Import required libraries
import pandas as pd
import numpy as np
from scipy import stats
from scipy.stats import poisson
#Read and load dataset
car=pd.read_excel('/content/drive/MyDrive/Data Analytics with python/P_distribution.xlsx')
car
```

```python
observed_freq=car['Frequency'] #Observed frequency
total_arrival=600
total_time_period=100
mu=total_arrival/total_time_period

Expected_freq=[]
for i in range(len(observed_freq)):
  E_freq=100*poisson.pmf(i,mu)
  Expected_freq.append(E_freq)
Expected_freq #Expected Frequency

rounded_exp_freq=[round(element,2) for element in Expected_freq]
rounded_exp_freq

df=pd.DataFrame({'Observed Frequency':observed_freq,'Expected Frequency':rounded_exp_freq}) #New dataframe
df
```

```python
#Calculated Chi-Square and p-value
from scipy.stats import chisquare
chisquare(f_obs=observed_freq,f_exp=rounded_exp_freq)
```

Classification and Regression Tree (CART)

Example: Computer Purchase

```python
#Import Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Read and load dataset
cart=pd.read_excel('/content/drive/MyDrive/Data Analytics with python/CART.xlsx')
cart
```

```python
#Data Encoding
import sklearn
from sklearn.preprocessing import LabelEncoder

#Function-LabelEncoder()
le_age=LabelEncoder()
le_income=LabelEncoder()
le_student=LabelEncoder()
le_credit_rating=LabelEncoder()
le_buys_computer=LabelEncoder()

#Function-fit_transform()
cart['age_n']=le_age.fit_transform(cart['age'])
cart['income_n']=le_income.fit_transform(cart['income'])
cart['student_n']=le_student.fit_transform(cart['student'])
cart['credit_rating_n']=le_credit_rating.fit_transform(cart['credit_rating'])
cart['buys_computer_n']=le_buys_computer.fit_transform(cart['buys_computer'])
cart
```

```python
#Drop Columns
cart_new=cart.drop(['age', 'income', 'student', 'credit_rating', 'buys_computer'], axis='columns')
cart_new
```

```python
#Dependent and Independent Variable selection
x=cart_new.drop(['buys_computer_n','RID'], axis='columns')
y=cart_new['buys_computer_n']
```

```python
#Build the Decision Tree Model without Splitting
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(x,y)
```

```python
#Visualishing Decision Tree
from sklearn.tree import export_graphviz
from io import StringIO
from IPython.display import Image
import pydotplus

dot_data=StringIO()
export_graphviz(model,out_file=dot_data,filled=True,rounded=True,special_characters=True,feature_names=x.columns,class_names=['0','1'])
graph=pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png('buys_computer.png')
```

```python
Image(graph.create_png())
```

Now splitting the data into train and test and checking the result

```python
#Splitting dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
#Build the decision tree
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
#Evaluating the model
from sklearn.metrics import accuracy_score
y_predict=model.predict(x_test)
accuracy_score(y_test,y_predict)
```

```python
#Visualishing Decision Tree
from sklearn.tree import export_graphviz
from io import StringIO
from IPython.display import Image
import pydotplus

dot_data=StringIO()
export_graphviz(model,out_file=dot_data,filled=True,rounded=True,special_characters=True,feature_names=x.columns,class_names=['0','1'])
graph=pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png('buys_computer.png')
```
